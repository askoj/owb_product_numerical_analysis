# Numerical Analysis Of The OWB Product

Prior to reading, this analysis assumes you are familiar with the following works:
- [1] _Acin, A., Fritz, T., Leverrier, A., and Sainz, A. B. (2015). A combinatorial approach to nonlocality and contextuality. Communications in Mathematical Physics, 334(2):533–628._
- [2] _Sainz, Ana Belén, and Elie Wolfe. "Multipartite composition of contextuality scenarios." Foundations of Physics 48.8 (2018): 925-953._
- [3] _The thesis of Abdul Karim Obeid, submitted to Queensland University of Technology in 2021_

This iPython Jupyter Notebook serves as an interactive and clarified method through which a numerical analysis can be undertaken on the 'Obeid Wolfe Bruza' (OWB) product [3], by comparison of its summated hyperedges to the theoretical minimum number of hyperedges of any compositional product capable of fulfilling the 'No-Disturbance' (ND) or 'No-Signalling' (NS) condition.

By application of this analysis, it has been found that any configuration ranging 2-9 parties, on 2-9 measurements, on 2-9 outcomes produces a result that is also the theoretical minimum number of hyperedges for any compositional product. Thus, it is speculated that this extends to any arbitrary parametric. Furthermore, this analysis is supplemented with the logs corresponding to the "Grid Sweep" test for all foredescribed parametrics, to further substantiate the validity of the OWB product (see `grid_sweep.txt`).

## Numerically Verifying the OWB Product By A Row-Reduced Echelon Form Matrix

To calculate the OWB product of a system, we can call the function `prod_OWB_edges`:

````python

from numerical_analysis import *

'''
    Determining the number of hyperedges in the OWB product for a system of:
        - three parties
        - bearing two measurements each
        - with two outcomes for each measurement
'''
prod_OWB_edges(3,2,2)
````

A LaTeX description of how this is calculated follows with the function `latex_owb_product`:

````python

latex_owb_product(3,2,2)
````

As shown above, the OWB product produces 38 hyperedges. This is the same number of hyperedges that define the theoretical lower bound necessary to convey either the No-Signalling or No-Disturbance condition. This is known by Sainz and Wolfe [2], who define the value as the number of rows in a row-reduced echelon form matrix:

${{\texttt {rref}}} \left( M_{\text {Normalization}}^{\vec {\mathbf{1}}} \cup M_{\text {NoSignalling}}^{\vec {\mathbf{0}}}\right)$

_See "Linear Algebra of Hypergraphs" in [2] for more details_

Here, I provide the function `number_of_rows_in_rref_matrix_for_B322` that applies their method to also determine the theoretical minimum number of hyperedges for a system (of three parties, bearing two measurements each, with two outcomes for each measurement) as calculated for the OWB product above:

````python

# Using the method of Sainz and Wolfe [2] to determine the number of hyperedges in a B322 system
number_of_rows_in_rref_matrix_for_B322()
````

As expected, the theoretical minimum is exactly equivalent to the number of hyperedges in the OWB product. In fact, this equality can be numerically evaluated for any configuration. In communications with Wolfe, there is an expression that determines the number of rows in any arbitrary `rref` function. By application of said expression, the function `print_owb_product_grid_sweep` allows for a grid sweep of all parametrics (i.e. the number of parties, measurements, and outcomes):

````python

'''
    This function will evaluate the hyperedges of the OWB product for all configurations in range 
    of 2-3 parties, 2 measurements, and 2 outcomes.
'''
print_owb_product_grid_sweep(3,2,2)
````
